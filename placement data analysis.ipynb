```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("D:/Placement_Data_Full_Class.csv")
```

```python
data.head()
```

```python
#pre-processing data
data.isnull().sum()
```

```python
data.gender.value_counts()  #does gender affects placements
```

```python
sns.countplot("gender", hue="status", data=data) #Efect of gender on placement
plt.show()
```

```python
data.groupby(['gender','status'])['status'].count().plot.line() #effect of gender on placement
```

```python
sns.kdeplot(data.salary[ data.gender=="M"])
sns.kdeplot(data.salary[ data.gender=="F"])
plt.legend(["Male", "Female"])
plt.xlabel("Salary (100k)")
plt.show()#relation between salary and gender
```

```python
plt.figure(figsize =(18,6))
sns.boxplot("salary", "gender", data=data)
plt.show()# relation of salary and gender
```
